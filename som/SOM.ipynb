```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
%matplotlib inline
from minisom import MiniSom
from sklearn.preprocessing import MinMaxScaler,scale
```

```python
data = pd.read_csv('~/Desktop/Concrete-NN/data/ets.csv')
data = data.drop('Cost ($/m3)',1)
scaler = MinMaxScaler(feature_range = (0,1))
scaled_data = scaler.fit_transform(data)
column_names = list(data.columns.values)[:-1]
data.head()
```

```python
SIZE = 5
som = MiniSom(SIZE,SIZE, len(scaled_data[0]), neighborhood_function='gaussian',
              sigma = 2, random_seed=10, learning_rate = 0.4)
som.pca_weights_init(scaled_data)
q_error_pca_init = []
iter_x = []
max_iter = 20000
for i in range(max_iter):
    percent = 100*(i+1)/max_iter
    rand_i = np.random.randint(len(scaled_data))
    som.update(scaled_data[rand_i], som.winner(scaled_data[rand_i]), i, max_iter)
    if (i+1) % 1000 == 0:
        error = som.quantization_error(scaled_data)
        q_error_pca_init.append(error)
        iter_x.append(i)
        print(f'\riteration={i:2d} status={percent:0.2f}% error={error}')


plt.plot(iter_x, q_error_pca_init)
plt.ylabel('quantization error')
plt.xlabel('iteration index')
```

```python
weights = som.get_weights()
plt.figure(figsize = (SIZE,SIZE))
for i, f in enumerate(column_names):
    plt.subplot(3,3, i+1)
    plt.title(f)
    plt.pcolor(weights[:,:,i].T, cmap='magma')
    plt.xticks(np.arange(SIZE+1))
    plt.yticks(np.arange(SIZE+1))
plt.colorbar()
plt.tight_layout()
plt.show()

```

```python
cat_map = som.labels_map(scaled_data,column_names)
plt.figure(figsize=(14,14))
for p, category in cat_map.items():
    category = list(category)
    x = p[0] + .1
    y = p[1] - .3
    for i, c in enumerate(category):
        off_set = (i+1)/len(category) - 0.05
        plt.text(x, y+off_set, c, fontsize=10)
plt.pcolor(som.distance_map().T, cmap='gray_r', alpha=.2)
plt.xticks(np.arange(SIZE+1))
plt.yticks(np.arange(SIZE+1))
plt.grid()

plt.show()
```
