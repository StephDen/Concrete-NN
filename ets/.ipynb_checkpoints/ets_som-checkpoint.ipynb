# SOM to map ingredient strength correlation

```python
#Import required libraries
import matplotlib.pylab as plt
%matplotlib inline
import pandas as pd
import numpy as np
import sompy
from sompy.visualiziation.mapview import View2D
```

```python
data = pd.read_csv('~/Desktop/Concrete-NN/data/ets.csv')
```

```python
data.describe()
```

```python
mapsize = [30,30]
som = sompy.SOMFactory.build(data,mapsize,mask = None, mapshape = 'planar',lattice = 'rect',normalization = 'var', initialization = 'pca',neighborhood='gaussian',training= 'batch', name='ETS DATA')
som.train(n_job=1,verbose = 'info')
```

```python
test = som.codebook.matrix
test = np.swapaxes(test,0,1)
som.codebook.matrix = test
som.codebook.matrix.shape
```

```python
from sompy.visualization.mapview import View2DPacked
v = View2DPacked(30,30,'test',text_size = 8)
v.show(som,col_sz=5,which_dim='all')
plt.show()
```
